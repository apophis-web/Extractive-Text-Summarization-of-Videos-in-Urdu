```python
import scipy.io.wavfile
import pandas
import nltk
import numpy
from collections import Counter
import math
import numpy
import speech_recognition as sr
from pydub import AudioSegment
from operator import itemgetter
from translate import Translator
from transformers import pipeline
```

```python
#Storing the audio in desired format and transcribing it in chunks of 60,000 milliseconds
r = sr.Recognizer()
transcript = ""
unit = 60000
sound = AudioSegment.from_mp3("voice.mp3") 
last_chunk_length = len(sound) % 60000
for i in range(len(sound) // unit):
  sound = AudioSegment.from_mp3("voice.mp3") 
  if (i == 0):  
    sound = sound[:unit]
    sound.export("audio.wav", format = "wav")
    audio = sr.AudioFile("audio.wav")
    with audio as source:
        audiodata = r.record(audio)
    query = r.recognize_google(audiodata, language = 'ur-PK')
    transcript = transcript + query
    print(query)
    print("\n")
  else:
    sound = sound[unit:(unit + 60000)]
    sound.export("audio.wav", format = "wav")
    audio = sr.AudioFile("audio.wav")
    with audio as source:
        audiodata = r.record(audio)
    query = r.recognize_google(audiodata, language = 'ur-PK')
    transcript = transcript + query
    print(query)
    print("\n")
  unit = unit + 60000
```

```python
#Removing special characters from the corpus and tokenizing it all into a list
special_characters = ['‘','٪','%', '!', '`', '"', ')', '(',"''", '.', ':','’’', "'", '"', '؟','‘','’','‘','،','“','’','•','|']
tokenized_words = []
for i in lines.split():
    word = i.split()
    if len(word) != 0:
        if word[0] not in special_characters:
                tokenized_words.append(i)
```

```python
def number_of_words(x):
    num = 0
    for i in x:
        num = num + 1
    return num
number_words = number_of_words(tokenized_words)
```

```python
#Creating a unigram from the corpus to rank Urdu language words according to their occurence probabilities
corpus_unigram = {}
temp = []
for index, i in enumerate(tokenized_words):
    if (index < number_words - 1):
        temp.append(i)
        temp = ' '.join(map(str, temp))
        if temp in corpus_unigram.keys():
            corpus_unigram[temp] += 1
        else:
            corpus_unigram[temp] = 1 
        temp = []
corpus_unigram
```

```python
corpus_unigram = sorted(corpus_unigram.items(), key = lambda x:x[1], reverse = True)
corpus_unigram = dict(corpus_unigram)
```

# Title Similarity Text Rank

```python
syns = {}
f = open("SYNS.txt", "r", encoding='utf-8')
f = f.readlines()

for i in range(len(f)):
  f[i] = f[i].replace("\n", "")

for i in range(len(f)):
  f[i] = f[i].split(":")
  key = f[i][0] 
  word = f[i][1]
  syns[key] = word 

import googletrans
from googletrans import Translator
def Trans(word):
  translator = Translator()
  ar = translator.translate(word, dest='ur').text
  return ar
```

```python
special_characters = ['‘','٪','%', '!', '`', '"', ')', '(',"''", '.', ':','’’', "'", '"', '؟','‘','’','‘','،','“','’','•','|']

title = "Evolution vs. Islam | Mooroo Ke Khyalaat"
title = title.split(" ")
for i in special_characters:
  try:
    title.remove(i)
  except:
    pass
for i in range(len(title)):
  try:
    title[i] = Trans(title[i])
  except:
    pass
```

```python
scores = {}
def CheckSynsinSynonyms(sent_word, tit_word):
  try:
    tit_syns = syns[tit_word]
    tit_syns = tit_syns.split(",")
    for i in tit_syns:
      if (i in syns[sent_word].split(",")):
        return True
  except:
    pass
  try:
    sent_syns = syns[sent_word]
    sent_syns = sent_syns.split(",")
    for i in sent_syns:
      if (i in syns[tit_word].split(",")):
        return True
  except:
    pass
  return False


for i in range(len(clean_relevant_sentences)):
  count = 0
  for sentence_word in clean_relevant_sentences[i]:
    for title_word in title:
      try:
        if (sentence_word == title_word):
          count = count + 3
      except:
        pass

      try:
        if sentence_word in syns[title_word]:
          count = count + 2
      except:
        pass
      if (CheckSynsinSynonyms(sentence_word, title_word) == True):
          count= count + 1

      try:
        if title_word in syns[sentence_word]:
          count = count + 2
      except:
        pass
  count = count / len(clean_relevant_sentences)
  scores[i] = count
```

```python
res = dict(sorted(scores.items(), key = itemgetter(1), reverse = True)[:math.ceil(len(relevant_sentences)* 0.2)])
result = {}
for i in res.keys():
  result[relevant_sentences[i]] = res[i]
```
